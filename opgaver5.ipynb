```python
from sympy import *
import scipy.signal as signal
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
init_printing()
j = I
```

```python
t = symbols('t', real=True)
omega = symbols('omega', real=True)
display(integrate(DiracDelta(t)*exp(-j*omega*t), (t, -oo, oo)))
display(1/(2*pi)*integrate(DiracDelta(t)*exp(-j*omega*t), (t, -oo, oo)))
display(1/(2*pi)*integrate(DiracDelta(t)*exp(-j*omega*t), (t, -oo, oo)))
```

```python
xe = exp(-4*abs(t))
xo = -exp(-4*abs(t))*Heaviside(-t)+exp(-4*abs(t))*Heaviside(t)
display(FourierTransform(xe, t, omega).doit())
display(FourierTransform(xo, t, omega).doit())
```

```python
R,C = symbols('R C')
H = (1/(R*C))/(j*omega+(1/(R*C)))
H
```

```python
gain = 20*log(abs(H.subs({R:1e3,C:1e-6})),10)
phase = arg(H.subs({R:1e3,C:1e-6}))
```

```python
#plot gain and phase a function of omega with a logarithmic omega axis
f = Symbol('f')
plot(gain.subs(omega,2*pi*f),(f,1,1e4),xscale='log',yscale='linear',ylim=(-40,0),axis_center=(1,0),xlabel='f',ylabel='dB',grid=True)
plot(180/pi*phase.subs(omega,2*pi*f),(f,1,1e4),xscale='log',yscale='linear',ylim=(-90,0),axis_center=(1,0),xlabel='f',ylabel='phase',grid=True)
```
