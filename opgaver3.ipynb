```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
init_printing()
from matplotlib import style

j = I
```

```python
#lp
t = Symbol('t', real=True,nonnegative=True)
s = Symbol('s')
x = Function('x')(t)
y = Function('y')(t)

a0 = 3.95e5
a1 = 2.83e3
b0 = 7.89e5
b1,b2 = 0,0

P = (s**2+a1*s+a0)
Q = (b0+b1*s+b2*s**2)

H = Q/P

rts = solve(P,s)
rts
```

```python
omega = Symbol('omega', real=True)
f = Symbol('f', real=True)
M = abs(H.subs(s,j*f*2*pi))
M_db = 20*log(M,10)
plot(M_db,(f,1,1e4),xscale='log',yscale='linear',ylim=(-80,40),axis_center=(1,0),xlabel='omega',ylabel='dB',grid=True);
```

```python
t,tau = symbols('t tau', real=True)
def step(offset):
    return Piecewise((0, t-offset<0), (1, True))

h = 311.8*(exp(rts[0]*t)+exp(rts[1]*t))

x = (2*step(0)).simplify().subs(t,tau)

y = integrate(h.subs(t,t-tau)*x, (tau,0,t)).simplify()
y
```

```python
t,tau = symbols('t tau', real=True)
def step(offset):
    return Piecewise((0, t-offset<0), (1, True))

h = -0.6283*(2-0.31415*t)*exp(-0.31415*t)

x = (2*step(0)).simplify().subs(t,tau)

y = 2*x.subs(tau,t) + integrate(h.subs(t,t-tau)*x, (tau,0,t))
plot(y,(t,0,50),xlabel='t',ylabel='y(t)',ylim=(-1,4),grid=True);
```
